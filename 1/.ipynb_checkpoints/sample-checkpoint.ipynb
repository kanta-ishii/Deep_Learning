```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
```

```python
#配列
x = np.array([1.0, 2.0, 3.0])
print(x)
type(x)
```

```python
#算術
x = np.array([1.0, 2.0, 3.0])
y = np.array([4.0, 5.0, 6.0])
```

```python
#ブロードキャスト
print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(x / 2.)
```

```python
A = np.array([[1, 2], [3, 4]])
print(A)
print(A.shape) #個数
print(A.dtype) #要素データ型
print(A.flatten()) #フラットに
```

```python
x = np.arange(0., 6., 0.1) #0から6まで0.1刻み生成
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0., 6., 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label = "sin")
plt.plot(x, y2, label = "cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title('sin & cos')
plt.legend()
plt.show()
```
